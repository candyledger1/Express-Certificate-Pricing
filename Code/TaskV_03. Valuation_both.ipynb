```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
# Load fixed volatility data
df_fixed = pd.read_csv("fixed_volatility_simulations.csv", parse_dates=["Date"])

# Load rolling volatility data
df_rolling = pd.read_csv("rolling_volatility_simulations.csv", parse_dates=["Date"])
```

```python
print("Fixed Volatility:")
print(df_fixed.head())

print("\nRolling Volatility:")
print(df_rolling.head())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Fixed volatility
axes[0].plot(df_fixed["Date"], df_fixed["MarketPrice"], label="Market", color="red")
axes[0].plot(df_fixed["Date"], df_fixed["Simulated_1Y_FixedVol"], label="1Y Fixed", color="blue")
axes[0].plot(df_fixed["Date"], df_fixed["Simulated_2Y_FixedVol"], label="2Y Fixed", color="green")
axes[0].set_title("Fixed Volatility")
axes[0].set_xlabel("Date")
axes[0].set_ylabel("Certificate Price (EUR)")
axes[0].grid(True)
axes[0].legend()

# Rolling volatility
axes[1].plot(df_rolling["Date"], df_rolling["MarketPrice"], label="Market", color="red")
axes[1].plot(df_rolling["Date"], df_rolling["Simulated_1Y_RollingVol"], label="1Y Rolling", color="blue")
axes[1].plot(df_rolling["Date"], df_rolling["Simulated_2Y_RollingVol"], label="2Y Rolling", color="green")
axes[1].set_title("Rolling Volatility")
axes[1].set_xlabel("Date")
axes[1].grid(True)
axes[1].legend()

plt.tight_layout(rect=[0, 0, 1, 0.99])  # Leave space for the supertitle
#fig.savefig("valuation.pdf", format='pdf', dpi=300, bbox_inches='tight')
plt.show()
```
